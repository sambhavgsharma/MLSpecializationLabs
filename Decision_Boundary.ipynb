```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1]).reshape(-1,1)
```

```python
fig, ax = plt.subplots(1,1,figsize=(4,4))
ax.scatter(x[:,0], x[:,1], c=y, cmap='Paired')
ax.axis([0, 4, 0, 3.5])
ax.set_ylabel('$x_1$')
ax.set_xlabel('$x_0$')
plt.show()
```

```python
def sigmoid(z):
  g = 1/(1+np.exp(-z))
  return g
def draw_vthresh(ax, x_value, left_color='blue', right_color='red', alpha=0.1, line_color='k', line_style='--'):
  y_min, y_max = ax.get_ylim()
  ax.axvspan(ax.get_xlim()[0], x_value, color=left_color, alpha=alpha)
  ax.axvspan(x_value, ax.get_xlim()[1], color=right_color, alpha=alpha)
  ax.axvline(x=x_value, color=line_color, linestyle=line_style)

z = np.arange(-10,11)
fig,ax = plt.subplots(1,1,figsize=(5,3))
ax.plot(z, sigmoid(z), c="b")
ax.set_title("Sigmoid function")
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')
draw_vthresh(ax,0)
```

```python
x0 = np.arange(0,6)
x1 = 3-x0
fig,ax = plt.subplots(1,1,figsize=(5,4))
ax.plot(x0,x1,c='g')
ax.axis([0,4,0,3.5])
ax.fill_between(x0,x1,alpha=0.2) #Decision Boundary Plotting
ax.scatter(x[:,0], x[:,1], c=y, cmap='Paired', marker='x')
ax.set_ylabel(r'$x_1$')
ax.set_xlabel(r'$x_0$')
plt.show()
```
