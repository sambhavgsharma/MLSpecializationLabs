```python
import numpy as np
import time
```

```python
# Vector Creation - Creating a Vector with Zeros
a = np.zeros(4);
print(a, a.shape, a.dtype)
```

```python
# Vector Creation -  Creating a vector with random values (Float)
a = np.random.random_sample(4)
print(a, a.shape, a.dtype)
```

```python
# Vector Creation -  Creating a vector with random values (Integers)
# a = np.random.random_integers(4) - Is deprecated
a = np.random.randint(1, 4+1, size=4) #Include size, if not then it returns a random number
print(a, a.shape, a.dtype)
```

```python
# Vector Creation -  Creating a vector with the first n whole numbers
a = np.arange(4)
print(a, a.shape, a.dtype)
```

```python
# Vector Creation -  Creating a vector manually
a = np.array([1, 4, 2, 4, 5])
print(a, a.shape, a.dtype)
a = np.array([1, 4, 2, 4, 5.]) #to make into float - enter float point entries
print(a, a.shape, a.dtype)
```

```python
# Indexing in Numpy array

a = np.arange(10) #0,1,2 . . 9
print(a)

# Accessing the first element
print(a[0])

# Accessing the last element
print(a[9]) #This is method 1 : Note not 10 .. as indexing starts from 0
print(a[-1]) #This is method 2
```

```python
# Vector Slicing

a = np.arange(10)
print(a)

# [start, stop, step] - start and stop are indices, step is an integer
c = a[1:8:1]
print(c)

c = a[2:7:2]
print(c)

# Accessing all elements after 3
c = a[3:]
print(c)

#Acessing all element  till 3
c = a[:3]
print(c)

# Acessing all elements
c = a[:]
print(c)
```

```python
a = np.array([1,2,3,4])
print(f"a             : {a}")
# negate elements of a
b = -a
print(f"b = -a        : {b}")

# sum all elements of a, returns a scalar
b = np.sum(a)
print(f"b = np.sum(a) : {b}")

b = np.mean(a)
print(f"b = np.mean(a): {b}")

b = a**2
print(f"b = a**2      : {b}")
```

```python
a = np.array([ 1, 2, 3, 4])
b = np.array([-1,-2, 3, 4])
print(f"Binary operators work element wise: {a + b}")
```

```python
a = np.array([1, 2, 3, 4])

# multiply a by a scalar
b = 5 * a
print(f"b = 5 * a : {b}")
```
