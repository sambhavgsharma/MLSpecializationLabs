```python
arr = np.arange(10)
print(arr)
for i in range(len(arr)):
  if i+1==1:
    sux="st"
  elif i+1==2:
    sux = "nd"
  elif i+1==3:
    sux = "rd"
  else:
    sux = "th"
  print(f"{arr[i]} is the {i+1}{sux} element")
```

```python
# Array Slicing
arr[0:9:2] #Even
arr[1:9:2] #Odd
arr[0:9:4]
```

```python
a = np.array([1, 2, 3, 4])
print(f"a(original) : {a}")
b = -a #Negattion
print(f"b(Negation) : {b}")
sum = np.sum(a) #returns a scalar
print(f"sum : {sum}")
c = a**2
print(f"c(Squares) : {c}")
```

```python
# Vector Ops
#   Addition and Substraction
a = np.array([[1,2,3,4],[4,3,2,1]])
b=  np.array([[2,4,6,8],[1,3,5,7]])

print(f"Add : {a+b}")
print(f"Substract: {a-b}")
```

```python
# Scalar Multiply
print(a*4) #Constant Multiply
```

```python
# Vector Product (Dot Product)
print(np.dot(a,b.T)) #Transpose of B (2,4) (4,2)
```
